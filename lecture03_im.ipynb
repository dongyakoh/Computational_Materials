# Incomplete Markets

We consider an economy where agents do not have an access to a complete set of contingent
claims. One of such extreme cases is self-insurance economy: agents can only save through
risk-free IOU.

### Permanent Income and Life-Cycle Hypothesis (PILCH)

Now let's think about the stochastic income. From the asset pricing lecture, we know that
FOC with respect to a risk-free asset $a_{t+1}$ will be

\begin{equation}
\lambda(s^t) = (1 + r)\sum_{s^{t+1}|s^t}\lambda(s^{t+1})
\end{equation}
where $\lambda$ indicates marginal utility of consumption. Then, formally writing
\begin{align*}
U'(c_t(s^t))  &= \beta (1+r)\sum_{s^{t+1}|s^t} \pi_{t+1}(s^{t+1}|s^t) U'(c_{t+1}(s^{t+1})) \\
&= \frac{1+r}{1+\rho} E_t U'(c_{t+1}(s^{t+1})) \\
\end{align*}
This is a standard stochastic Euler equation for PILCH type models. If $\rho = r$ (implying that the subjective discount rate is equal to the objective discount rate), then the marginal utility of consumption follows a martingale, i.e. its period $t$ marginal utility is a good predictor of $t+1$ marginal utility. 



### Empirical Test of PILCH

Assume that the period utility function is quadratic and separable:
\begin{equation*}
U(c_t(s^t)) = -\frac{1}{2} (c_t(s^t) - \bar{c})^2
\end{equation*}
where $\bar{c}$ is a subsistence level of consumption. Then the Euler equation becomes
\begin{equation*}
E_t c_{t+1} = \alpha_1 + \alpha_2 c_t
\end{equation*}
where $\alpha_1 = \bar{c}\left( 1 - \frac{1+\rho}{1+r} \right)$ and $\alpha_2 = \frac{1+\rho}{1+r}$.
\item With $\rho=r$, then
\begin{equation*}
E_t c_{t+1} = c_t
\end{equation*}
that is, consumption follows a martingale.


The agent's optimal consumption decisions obey certainty equivalence; comparing the rules for optimally allocating consumption over time in the certainty and the uncertainty case. Consumption should obey the regression,
\begin{equation*}
c_{t+1} = \alpha_1 + \alpha_2 c_t + u_{t+1}
\end{equation*}
where $u_{t+1}$ is a random variable with $E_t u_{t+1} = 0$. This equation implies that the consumption at time $t+1$ is perfectly predicted by period $t$ consumption and no other variables that are in the household's information set (e.g. $y_t$, $a_{t}$, or $c_{t-1}$) at period $t$ should help predict it.


The income realization in period $y_{t+1}$ does affect $c_{t+1}$, but only the part that is unexpected change from income in period $t$. Remember in complete market, not even unexpected changes in income should affect consumption. The two implications above were empirically tested first in a seminal paper by Robert Hall (1978), running two regressions with "aggregate data".
\begin{align*}
c_{t+1} &= \alpha_1 + \alpha_2 c_t + \alpha_3 c_{t-1} + \alpha_4 c_{t-2} + \alpha_5 c_{t-3} + u_{t+1}\\
c_{t+1} &= \beta_1 + \beta_2 c_t + \beta_3 y_{t}+ v_{t+1}
\end{align*}


If the martingale hypothesis is true, then we expect $\alpha_3 = \alpha_4 = \alpha_5 = \beta_3 = 0$. The test results tell us that the hypothesis cannot be rejected. This is strong evidence in favor of the martingale hypothesis. However, when Hall included current stock prices as a proxy for current wealth into the regression, he finds that they are statistically significant, formally rejecting the martingale hypothesis. Hence he concludes that this hypothesis is rejected by the data.

### Consumption responses to income shocks under certainty equivalence

Using the sequential budget constraint, we can construct lifetime budget constraint as before. In a similar way, the lifetime budget in a stochastic setting becomes
\begin{equation*}
E_t \sum_{j=0}^{T-t} \frac{c_{t+j}}{(1+r)^j} = E_t \sum_{j=0}^{T-t} \frac{y_{t+j}}{(1+r)^j} + a_t \equiv W_t
\end{equation*}
Now, with a quadratic preference we had $E_t c_{t+1} = c_t$. Substitute this in to the lifetime budget constraint we have
\begin{equation*}
c_t  \sum_{j=0}^{T-t} \frac{1}{(1+r)^j} = c_t  \frac{(1+r)(1-(1+r)^{-(T-t+1)})}{r} = W_t
\end{equation*}
Therefore, the consumption  will be
\begin{equation*}
c_t = \theta^{-1}_t\frac{r}{1+r} W_t
\end{equation*}
where $\theta_t = (1-(1+r)^{-(T-t+1)})$ and if the time horizon is infinite,
\begin{equation*}
c_t = \frac{r}{1+r} W_t
\end{equation*}


* This equation determines the household's current consumption. It is comparable to Keynes' marginal propensity to consume. Keynes believe that the current consumption is a function of current disposable income. But this has been refuted by an empirical test. PILCH model tells us that the current consumption is determined as an annuity value of future expected income and wealth.

* So far, we haven't assumed any borrowing limits. In other words, households can borrow as much as they want. Then, a natural question is: why wouldn't households perfectly smooth their consumption? The current consumption equation and the Euler equation in fact answer this question. First, the consumption is ex-ante constant over time (consumption follows martingale). However, the current consumption is determined by the future expected income and wealth, the ex-post consumption deviates from the constant consumption level given any shocks to income. 



Let's rewrite this equation in a way that we link stochastic income to the changes in consumption.
\begin{align*}
\theta_t c_t &= \frac{r}{1+r} W_t =   \frac{r}{1+r}a_t + \frac{r}{1+r}E_t \sum_{j=0}^{T-t} \frac{y_{t+j}}{(1+r)^j}\\
&= r (a_{t-1} + y_{t-1} - c_{t-1}) + \frac{r}{1+r}E_t \sum_{j=0}^{T-t} \frac{y_{t+j}}{(1+r)^j}\\
\frac{\theta_t c_t + rc_{t-1}}{1+r}&= \frac{r}{1+r} (a_{t-1} + y_{t-1}) + \frac{r}{(1+r)^2}E_t \sum_{j=0}^{T-t} \frac{y_{t+j}}{(1+r)^j}
\end{align*}
Subtracting $\theta_{t-1} c_{t-1}$ on both sides, we have
\begin{align*}
\frac{\theta_t c_t + rc_{t-1}}{1+r} - \theta_{t-1} c_{t-1} &= \frac{r}{1+r} (a_{t-1} + y_{t-1}) + \frac{r}{(1+r)^2}E_t \sum_{j=0}^{T-t} \frac{y_{t+j}}{(1+r)^j} \\
&- \frac{r}{1+r}a_{t-1} -  \frac{r}{1+r} E_{t-1} \sum_{j=0}^{T-t+1} \frac{y_{t+j-1}}{(1+r)^j}\\
&= \frac{r}{1+r} y_{t-1} + \frac{r}{(1+r)^2}E_t \sum_{j=0}^{T-t} \frac{y_{t+j}}{(1+r)^j}  -  \frac{r}{1+r} E_{t-1} \sum_{j=0}^{T-t+1} \frac{y_{t+j-1}}{(1+r)^j}
\end{align*}
Rearranging the left hand side gives
\begin{align*}
\frac{\theta_t c_t + rc_{t-1}}{1+r} - \theta_{t-1} c_{t-1} &= \frac{\theta_t \Delta c_t}{1+r}
\end{align*}
Then we have 
\begin{align*}
\theta_t \Delta c_t &= r y_{t-1} + \frac{r}{1+r}E_t \sum_{j=0}^{T-t} \frac{y_{t+j}}{(1+r)^j}  -  r E_{t-1} \sum_{j=0}^{T-t+1} \frac{y_{t+j-1}}{(1+r)^j} \\
&= \frac{r}{1+r}E_t \sum_{j=0}^{T-t} \frac{y_{t+j}}{(1+r)^j}  -  r E_{t-1} \sum_{j=1}^{T-t+1} \frac{y_{t+j-1}}{(1+r)^j}\\
&= \frac{r}{1+r}E_t \sum_{j=0}^{T-t} \frac{y_{t+j}}{(1+r)^j}  -  \frac{r}{1+r} E_{t-1} \sum_{j=0}^{T-t} \frac{y_{t+j}}{(1+r)^j}\\
&= \frac{r}{1+r} \sum_{j=0}^{T-t} \frac{(E_t - E_{t-1})y_{t+j}}{(1+r)^j}\\
\end{align*}
This implies that the changes in consumption is the annuity value of innovation to future income between period $t$ and $t-1$.


Let's assume that the income process follows
\begin{align*}
y_t &= y^p_t + u_t \\
y^p_t &= y^p_{t-1} + v_t
\end{align*}
where $y^p_t$ is the permanent part of the income process, which follows unit root process, and $u_t$ is a transitory part and $(u_t,v_t)$ are iid shocks, independent from each other, and $E_t u_{t+j} = E_t v_{t+j} = 0$ for $j>0$. Then this can be written as
\begin{align*}
y^p_t &= y_t - u_t \\
y^p_t &= y^p_{t-1} + v_t \\
y_t &= y_{t-1} + u_t - u_{t-1} + v_t \\
y_{t+j} &= y_{t-1} + u_{t+j}  - u_{t-1} + \sum_{\tau=t}^{t+j} v_{\tau} 
\end{align*}


Then taking the expectation with time $t$ and $t-1$ information
\begin{align*}
E_t y_{t+j} &= y_{t-1} - u_{t-1} + v_{t} + u_t I(j=0)\\ 
E_{t-1} y_{t+j} &= y_{t-1} - u_{t-1} \\
(E_t - E_{t-1}) y_{t+j} &= v_{t} + u_t I(j=0)\\ 
\end{align*}


Plugging into the consumption response, we have
\begin{align*}
\theta_t \Delta c_t &= \frac{r}{(1+r)} \sum_{j=0}^{T-t} \frac{(E_t - E_{t-1})y_{t+j}}{(1+r)^j}\\
&= \frac{r}{(1+r)}u_t + \frac{r}{(1+r)} v_t \sum_{j=0}^{T-t} \frac{1}{(1+r)^j}\\
&= \frac{r}{(1+r)}u_t + \theta_t v_t\\
\Delta c_t &= \frac{r \theta_t^{-1}}{(1+r)}u_t + v_t
\end{align*}

This equation tells us that households adjust their consumption one for one with the permanent shock $v_t$. But only change their consumption mildly by $\frac{r \theta_t^{-1}}{(1+r)}$ as response to a temporary shock $u_t$. 


* This discussion assumes that households can distinguish permanent from transitory income shocks. But, let's suppose that households are unable to make this distinction, and let income take the form
\begin{align*}
y_t = y_{t-1} + \varepsilon_t - \gamma \varepsilon_{t-1}, \quad \gamma \in [0,1]
\end{align*}
If $\gamma = 1$, it corresponds to transitory shocks, while it corresponds to the permanent one  if $\gamma = 0$. If $\gamma \in (0,1)$, then the shock is somewhere in between but households cannot tell. Then
\begin{align*}
(E_t - E_{t-1}) y_{t+j} &= \varepsilon_{t} - \gamma \varepsilon_t I(j>0)
\end{align*}
and thus we have
\begin{align*}
\theta_t \Delta c_t &= \frac{r \gamma}{(1+r)}\varepsilon_t + (1-\gamma)\theta_t \varepsilon_t
\end{align*}
or
\begin{align*}
\Delta c_t &= \frac{r \gamma \theta^{-1}_t }{(1+r)}\varepsilon_t + (1-\gamma) \varepsilon_t
\end{align*}
The households try to decompose the new information about income contained in $\varepsilon_t$ into a component reflecting new information about the permanent component of income and into a component reflecting the transitory innovation. If $\gamma = 1$, the agent responds as to transitory shocks. If $\gamma = 0$, the agent responds as to permanent shocks.


* Finally, income process following a simple $AR(1)$ process
\begin{align*}
y_t = \rho y_{t-1} + \varepsilon_t
\end{align*}
then the response should lie in between a purely transitory shock $\rho=0$ and a permanent shock $\rho>0$.


### Precautionary Saving

We will work on the PILCH model in a partial equilibrium setting in this section. By partial equilibrium we mean the interest rate that the agent faces is exogenously taken. In this section, we will discuss about the conditions under which the household's consumption responds to the conditional variance of future labor income.

Starting from the stochastic Euler equation:
\begin{align*}
U'(c_t) &= \beta (1+r) E_t U'(c_{t+1})
\end{align*}

For simplicity, let's assume $\beta (1+r) = 1$. Consider a simple two-period model with period 0 income being $y_0$ and period 1 income being stochastic, $y_1 = \bar{y} + \varepsilon_1$ where $\varepsilon_1$ is an iid random variable in period 1 and $E_0 y_1 = \bar{y}$.

Then consumption at period 1 will be
\begin{align*}
c_1 = y_0 + \bar{y} - c_0 + \varepsilon_t \equiv w - c_0 + \varepsilon_t 
\end{align*}

Then the Euler equation can be written as
\begin{align*}
U'(c_0) &= E_0 U'(w - c_0 + \varepsilon_t)
\end{align*}

To explore as to how the consumption in period 0 varies with the degree of uncertainty about labor income in period 1, we assume

\begin{align*}
\varepsilon_t = \left\{
\begin{array}{cc}
-\eta & \text{with probability} \quad 1/2 \\ 
\eta & \text{with probability} \quad 1/2 \\ 
\end{array}
\right.
\end{align*}

so that $Var(\varepsilon_t) = \eta^2$. This setting implies that as $\eta$ becomes larger, the variance of the income shock becomes larger.


Now totally differentiate the Euler equation evaluated at $\eta=0$:
\begin{align*}
U''(c_0)\frac{dc_0}{d\eta} &=  \frac{1}{2} (U''(w - c_0(\eta) + \eta)\left( -\frac{dc_0}{d\eta} + 1 \right) + U''(w - c_0(\eta) - \eta)\left( -\frac{dc_0}{d\eta} - 1 \right))
\end{align*}

and solve for $\frac{dc_0}{d\eta}$, we have

\begin{align*}
\frac{dc_0}{d\eta} &=  \frac{\frac{1}{2} (U''(w - c_0(\eta) + \eta) - U''(w - c_0(\eta) - \eta))}{U''(c_0) + \frac{1}{2} (U''(w - c_0(\eta) + \eta) + U''(w - c_0(\eta) - \eta))} 
\end{align*}

Since assuming strictly concave utility function, the denominator is unambiguously negative. The numerator is positive if and only if
\begin{align*}
U''(w - c_0(\eta) + \eta) - U''(w - c_0(\eta) - \eta) > 0
\end{align*}

This is true for arbitrary $\eta>0$ if and only if $U'''(c)>0$. This implies that the consumption in period 0 declines in reaction to an increase in uncertainty about period 1 income if and only if the third derivative of the utility function is positive. Thus, a sufficient (and necessary) condition for precautionary saving (i.e. increased saving as reaction to increase uncertainty of future labor income) is strictly convex marginal utility. Kimball (1990) has defined the term "prudence" to mean the "the propensity to prepare and forearm oneself in the face of uncertainty" and hence the intensity of the precautionary saving motive. The absolute and relative prudence are defined as

\begin{align*}
p(c) &= -\frac{U'''(c)}{U''(c)}\\
pr(c) &= -\frac{cU'''(c)}{U''(c)}.
\end{align*}

Note that prudence and risk aversion are not the same thing; risk aversion is controlled by the concavity of the utility function whereas prudence is controlled by the convexity of the marginal utility function. The absolute and relative risk aversion are defined as

\begin{align*}
\sigma(c) &= -\frac{U''(c)}{U'(c)}\\
\sigma r(c) &= -\frac{cU''(c)}{U'(c)}
\end{align*}


**Parametric example:**

Assume CRRA utility and $\log(c_{t+1})$ is normally distributed with mean $\mu$ and variance $\sigma^2_c$. Then the stochastic Euler equation becomes

\begin{align*}
1 &= \beta (1+r) E_t \left( \frac{c_{t+1}}{c_t} \right)^{-\gamma} \\
&= E_t \exp ( \log\beta + \log(1+r) - \gamma \log c_{t+1} + \gamma \log c_t) \\
&= \exp ( \log\beta + \log(1+r)  + \gamma \log c_t )E_t \exp (- \gamma \log c_{t+1}) \\
&= \exp ( \log\beta + \log(1+r)  + \gamma \log c_t )\exp \left(- \gamma E_t \log c_{t+1} + + \frac{1}{2} \gamma^2 \sigma^2_c \right) \\
&= \exp \left( - \gamma E_t \Delta \log c_{t+1} + \log\beta + \log(1+r)  + \frac{1}{2} \gamma^2 \sigma^2_c \right) \\
\end{align*}

And by taking logs,

\begin{align*}
E_t \Delta \log c_{t+1}  = \frac{1}{\gamma }\left(\log\beta + \log(1+r)  + \frac{1}{2} \gamma^2 \sigma^2_c \right)
\end{align*}

Remember that without uncertainty

\begin{align*}
\Delta \log c_{t+1}  = \frac{1}{\gamma}\left( \log\beta + \log(1+r) \right)
\end{align*}

Due to the precautionary motive, uncertainty about future consumption tilts the consumption profile upward in expectation. That is households tend to postpone consumption for precautionary motives. The degree to which there is precautionary saving is determined by the parameter $\gamma$. All variables that help to predict the variability of future consumption $\sigma^2_c$ will help to predict expected consumption growth. For example, more consumption insurance implies lower $\sigma^2_c$ and hence lower future consumption variability.

### Liquidity Constraint

So far, we have assumed that the household can borrow up to some arbitrarily large amount, up to the no-Ponzi scheme condition which was assumed to be generous enough never to be binding. Now let us assume that there exists borrowing constraints and for simplicity, let's assume zero borrowing limit ($a_{t+1}(s^t)\geq 0$ for all $s^t$). These constraints may or may not be binding, but we have to take this constraint into account when deriving the stochastic Euler equation. When we attach Lagrange multiplier $\mu(s^t)$ to the borrowing constraint $a_{t+1}(s^t)\geq 0$ at every history $s^t$. Then the first order condition becomes

\begin{align*}
U'(c_t) &\geq \beta(1+r)E_t U'(c_{t+1}) \\
U'(c_t) &= \beta(1+r)E_t U'(c_{t+1}) \quad \text{if }a_{t+1}>0 
\end{align*}


This Euler equation, which sometimes holds with inequality, is the basis for empirical tests for liquidity constraints. Zeldes(1989) empirically tests whether current income helps forecast future consumption growth. As evidence for the presence of liquidity constraints would be interpreted a finding which shows that current income helps predict consumption growth for the low-wealth group, but not for the high-wealth group. Zeldes finds such evidence and concludes that borrowing constraints are important in shaping consumption choices, at least for the currently wealth-poor.


Note from the budget constraint
\begin{align*}
c_t &= y_t + a_t - \frac{a_{t+1}}{1+r} \\
&\leq y_t + a_t
\end{align*}

since $a_{t+1} \geq 0$. If $a_{t+1} = 0$, then

\begin{align*}
U'(c_t) = U'(y_t + a_t)
\end{align*}

and if $a_{t+1} > 0$, then

\begin{align*}
U'(y_t + a_t) < U'(c_t) = \beta(1+r)E_t U'(c_{t+1})
\end{align*}

Therefore, the Euler equation with borrowing limits can be compactly written as

\begin{align*}
U'(c_t) = \max \{ U'(y_t + a_t), \beta(1+r)E_t U'(c_{t+1})\}
\end{align*}


Note that even with certainty equivalence case, if the household is borrowing constrained in the future $s>0$, it is the case that $c_t < E_t c_{t+s}$. This implies that even if the liquidity constraint is not binding in period $t$, future potential borrowing constraints affect current consumption choices. Now, suppose the variance of future income increases (say for $t+1$), making lower realizations of $y_{t+1}$ possible. This makes a household more likely to be borrowing constrained in the future. As a result, saving increases in reaction to increases in uncertainty about future income. Note that we obtain precautionary savings behavior without prudence purely from the existence of liquidity constraints (and risk aversion). Therefore, precautionary motive may have a preference-based interpretation or an institution-based interpretation.

### Household Problem with Idiosyncratic Income Shocks

Infinitely lived households / consumers face idiosyncratic income shocks

A unit interval of  *ex ante* identical households face a common borrowing constraint

The savings problem faced by a typical  household is

$$
\max \mathbb E \sum_{t=0}^{\infty} \beta^t u(c_t)
$$

subject to

$$
a_{t+1} + c_t \leq w z_t + (1 + r) a_t
\quad
c_t \geq 0,
\quad \text{and} \quad
a_t \geq -B
$$

where

- $ c_t $ is current consumption  
- $ a_t $ is assets  
- $ z_t $ is an exogenous component of labor income capturing a stochastic income risk.  
- $ w $ is a wage rate  
- $ r $ is a net interest rate  
- $ B $ is the maximum amount that the agent is allowed to borrow  


The exogenous process $ \{z_t\} $ follows a finite state Markov chain with given stochastic matrix $ P $

The wage and interest rate are fixed over time

In this simple version of the model, households supply labor  inelastically because they do not value leisure.

In a recursive form:

$$
v(a,z) = \max_{c,a'} u(c) + \beta  \mathbb E  v(a',z')
$$

subject to

$$
c =  w z + (1 + r) a - a'
\quad
\text{and}
\quad
a \geq -B
$$



```python
import numpy as np

class household:
    # INITIALIZATION OF THE CLASS WITH GIVEN PARAMETERS VALUES
    def __init__(self,  na = 100,
                        nz = 2,
                        Π = [[0.9, 0.1], [0.1, 0.9]],  # Markov chain
                        zgrid = [0.1, 1.0],
                        r = 0.03,                    # interest rate
                        w = 0.986,                     # wages
                        β = 0.96,                    # discount factor
                        σ = 2,                       # discount factor
                        a_min = -2,
                        a_max = 10):


        # GRID FOR a and e
        self.na             = na
        self.nz             = nz
        self.agrid          = np.zeros(na)
        self.zgrid          = zgrid

        # PARAMETERS FOR UTILITY FUNCTION
        self.β              = β
        self.σ              = σ
        self.r              = r
        self.w              = w
        self.a_min          = a_min
        self.a_max          = a_max
        self.Π              = Π

        # INITIALIZE VALUE FUNCTION AND POLICY FUNCTION
        self.V              = np.zeros((na,nz))
        self.ap             = np.zeros((na,nz))


    def init_agrid(self):
        '''
        This function constructs and returns grid points for state k.
        The grid points are in equidistance.
        '''
        self.agrid = np.linspace(self.a_min,self.a_max,self.na)
        return self.agrid


    def util(self,cons):
        '''
        This function returns the value of CRRA utility with ssigma
        u(c) = c**(1-ssigma)/(1-ssigma)
        '''
        if self.σ != 1:
            uu = cons**(1-self.σ)/(1-self.σ)
        else:
            uu = np.log(cons)
        return uu


    def get_V(self):
        '''
        This function updates the value function
        '''
        return self.V

    def set_V(self,VV):
        '''
        This function updates the value function
        '''
        self.V   = VV
        
    def set_ap(self,ap):
        '''
        This function updates the policy function
        '''
        self.ap   = ap
```

```python
def bellman(ap,hh,a0,z0,VV,P):
    '''
    This function computes bellman equation for a given state (a0,e0).
    Input:
        ap: evaluating point
        hh: household object
        (a0,e0) state
        V1: value function at age t+1 evaluated at (a',e')
        P1: probability distribution of e' conditional on the current e0
    Output:
        -vv: bellman equation
    ''' 

    agrid   = hh.agrid
    w       = hh.w
    r       = hh.r
    β       = hh.β
    

    # Interpolate next period's value function evaluated at (a',e')
    # using 1-dimensional interpolation function in numpy
    V0      = np.interp(ap,agrid,VV[:,0])
    V1      = np.interp(ap,agrid,VV[:,1])
    EV      = P[0]*V0 + P[1]*V1
        
    # Interpolated value cannot be NaN or Inf
    if np.isnan(V0) or np.isinf(V0): print("bellman: V0 is NaN.")
    if np.isnan(V1) or np.isinf(V1): print("bellman: V1 is NaN.")

    # Compute consumption at a given (a0,e0) and a'       
    cons = w * z0 + (1 + r) * a0 - ap
    
    # Consumption must be non-negative
    if cons<=0:
        vv = (cons-10)*1000
    else:
        # Compute value function
        vv  = hh.util(cons) + β*EV
#    print(a0,e0,ap,cons,vv)
    
    return -vv
```

```python
from scipy.optimize import fminbound
import sys
import matplotlib.pyplot as plt

#if __name__ == "__main__":

na = 100
nz = 2
Π = [[0.9, 0.1], [0.1, 0.9]]  # Markov chain
zgrid = [0.1, 1.0]   
    
# Instantiate NGM object
hh = household(na,nz,Π,zgrid)
    
# Construct discretized state space in NGM object
agrid = hh.init_agrid()    


tol = 1e-5
diff = 1e5
niter  = 0

#Iterate until the difference of old and new value functions are within tol range.
while diff>tol:

    # Start with the current value function
    V_old   = hh.get_V()
    V_new   = np.zeros((na,nz))
    ap_new  = np.zeros((na,nz))

    
    # Solve for optimal capital and value function at each state k0
    for ind in range(na * nz):
        ia  = ind // nz
        iz  = ind % nz
                        
        a0  = agrid[ia]
        z0  = zgrid[iz]
        P   = Π[iz]
        ap_new[ia,iz] = fminbound(bellman,agrid[0],agrid[na-1],args=(hh,a0,z0,V_old,P))
        V_new[ia,iz] = -bellman(ap_new[ia,iz],hh,a0,z0,V_old,P)
    
    # Update the value function
    hh.set_ap(ap_new)
    hh.set_V(V_new)
    
    # Compute the difference between new and old value functions by sup norm
    diff = np.max(abs(V_old - V_new))
    niter += 1
    if niter%100==0:
        sys.stdout.write('Number of Iteration = %d, Difference = %f.\n' % (niter,diff))

sys.stdout.write('Value function converged.\n')

# Plot Policy Function
fig, ax = plt.subplots(figsize=(7, 5))
plt.plot(hh.agrid, hh.ap[:,0],label="bad shock")
plt.plot(hh.agrid, hh.ap[:,1],label="good shock")
plt.plot(hh.agrid, hh.agrid,'k--',label="$45^{\circ}$ line")
plt.xlabel("State space, a")
plt.ylabel("Policy function a' ")
plt.legend(loc='lower right', fontsize = 14)
plt.show()
```
